#### - Load and Explore the dataset:

#### - Clean and preprocess the raw data:

```python
#Check if dataset contain missing values
df.isnull().sum()
```

```python
#Check if data contain duplicated
df.duplicated().sum()
```

```python
# Remove duplicate entries
df_cleaned = df.drop_duplicates()
```

#### - Summarize important features:

```python
# Summarize insights
crime_counts = df_cleaned['Category'].value_counts().head(10).reset_index()
district_counts = df_cleaned['PdDistrict'].value_counts().reset_index()
resolution_counts = df_cleaned['Resolution'].value_counts().reset_index()

# Rename columns for better visualization
crime_counts.columns = ["Category", "Count"]
district_counts.columns = ["District", "Count"]
resolution_counts.columns = ["Resolution", "Count"]

# Set Seaborn style
sns.set_style("whitegrid")

# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Top 10 Crime Categories
sns.barplot(data=crime_counts, x="Count", y="Category", palette="Blues_r", ax=axes[0], hue="Category", legend=False)
axes[0].set_title("Top 10 Crime Categories")
axes[0].set_xlabel("Count")
axes[0].set_ylabel("Crime Category")

# Crime Distribution by District
sns.barplot(data=district_counts, x="Count", y="District", palette="Oranges_r", ax=axes[1], hue="District", legend=False)
axes[1].set_title("Crime Distribution by District")
axes[1].set_xlabel("Count")
axes[1].set_ylabel("District")

# Resolution Outcomes
sns.barplot(data=resolution_counts, x="Count", y="Resolution", palette="Greens_r", ax=axes[2], hue="Resolution", legend=False)
axes[2].set_title("Case Resolution Outcomes")
axes[2].set_xlabel("Count")
axes[2].set_ylabel("Resolution Type")

# Adjust layout and display
plt.tight_layout()
plt.show()
```

#### - Visualize trends:

```python
#Visualize trends

#Convert the Dates Column to DateTime Format
df['Dates'] = pd.to_datetime(df['Dates'])

#Plot Crime Frequency Over Time
# Resample by day and count occurrences
df_daily = df.resample('D', on='Dates').size()

# Plot the trend
plt.figure(figsize=(12, 6))
df_daily.plot()
plt.title('Crime Frequency Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
#Plot Crime Categories Over Time
# Group by Date and Category, then count occurrences
df_category_trends = df.groupby([df['Dates'].dt.date, 'Category']).size().unstack().fillna(0)

# Plot the trends
plt.figure(figsize=(12, 8))
df_category_trends.plot(kind='line', stacked=True, figsize=(14, 8))
plt.title('Crime Categories Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.legend(title='Crime Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
#Crime Frequency by Day of the Week
# Get the day of the week (0 = Monday, 6 = Sunday)
df['DayOfWeek'] = df['Dates'].dt.day_name()

# Group by DayOfWeek and count occurrences
df_day_of_week = df.groupby('DayOfWeek').size()

# Plot the frequency of crimes by day of the week
plt.figure(figsize=(8, 6))
df_day_of_week.plot(kind='bar', color='skyblue')
plt.title('Crime Frequency by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#Crime Locations by District
# Group by Police District and count occurrences
df_district = df.groupby('PdDistrict').size().sort_values(ascending=False)

# Plot the crime distribution by district
plt.figure(figsize=(12, 6))
df_district.plot(kind='bar', color='lightcoral')
plt.title('Crime Distribution by Police District')
plt.xlabel('Police District')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### 2)  Crime Classification & Severity Assignment

#### - Predicting Crime Type:

```python
# Map categories to numeric values
category_mapping = {category: idx for idx, category in enumerate(df['Category'].unique())}
y = df['Category'].map(category_mapping)
```

```python
# I used it in level 4
print(category_mapping)
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.pipeline import make_pipeline

X = df['Descript']
y = df['Category']

# Map categories to numeric values
category_mapping = {category: idx for idx, category in enumerate(df['Category'].unique())}
y = df['Category'].map(category_mapping)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Create a model pipeline with TfidfVectorizer and LogisticRegression
model = make_pipeline(TfidfVectorizer(), LogisticRegression())

# Step 4: Train the model
model.fit(X_train, y_train)

# Step 5: Predict crime categories on test data
y_pred = model.predict(X_test)

# Step 6: Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Map predicted categories to crime names 
predicted_categories = [list(category_mapping.keys())[list(category_mapping.values()).index(label)] for label in y_pred]
```

#### - Potential Improvements:

- Change Model: I can try different models such as Random Forest or XGBoost, which might perform better on complex data, and compare them using cross-validation.

- Tune Hyperparameters: I can optimize the Logistic Regression model by adjusting hyperparameters using GridSearchCV or RandomizedSearchCV for better performance.

- Cross-Validation: I can use k-fold cross-validation to assess the model's generalization and reduce overfitting from a single test split.

- Evaluate with Metrics: I can use precision, recall, F1-score, and confusion matrix in addition to accuracy to get a more detailed evaluation of the model’s performance across all classes.

##### In this code I changed model & add additional metrics:

```python
# Create a model pipeline with TfidfVectorizer and RandomForestClassifier
model = make_pipeline(TfidfVectorizer(), RandomForestClassifier(n_estimators=200, random_state=42))

#Train the model
model.fit(X_train, y_train)

#Predict crime categories on test data
y_pred = model.predict(X_test)
# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Additional Evaluation Metrics
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Map predicted categories to crime names 
predicted_categories = [list(category_mapping.keys())[list(category_mapping.values()).index(label)] for label in y_pred]
```

#### - Assigning Crime Severity:

```python
# Define the crime-to-severity mapping (mapping crime types to severity levels)
# I used it in level 4
crime_severity_mapping = {
    'NON-CRIMINAL': 1, 
    'SUSPICIOUS OCCURRENCE': 1, 
    'MISSING PERSON': 1, 
    'RUNAWAY': 1, 
    'RECOVERED VEHICLE': 1,
    'WARRANTS': 2, 
    'OTHER OFFENSES': 2, 
    'VANDALISM': 2, 
    'TRESPASS': 2, 
    'DISORDERLY CONDUCT': 2, 
    'BAD CHECKS': 2, 
    'LARCENY/THEFT': 3, 
    'VEHICLE THEFT': 3, 
    'FORGERY/COUNTERFEITING': 3, 
    'DRUG/NARCOTIC': 3, 
    'STOLEN PROPERTY': 3, 
    'FRAUD': 3, 
    'BRIBERY': 3, 
    'EMBEZZLEMENT': 3, 
    'ROBBERY': 4, 
    'WEAPON LAWS': 4, 
    'BURGLARY': 4, 
    'EXTORTION': 4, 
    'KIDNAPPING': 5, 
    'ARSON': 5
}

# Map the predicted crime categories to severity levels
predicted_severity = [crime_severity_mapping.get(crime, 'Unknown') for crime in predicted_categories]
```
